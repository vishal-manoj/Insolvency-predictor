```python
x = df.drop('Bankrupt?', axis=1)
y = df['Bankrupt?']
x
```

```python
plt.figure(figsize = (50,50), dpi=200)
sns.heatmap(x.corr(), annot=True)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
log_model = LogisticRegression(max_iter=1000)
random_forest_model = RandomForestClassifier()
decision_tree_model = DecisionTreeClassifier()
svc_model = SVC()
knn_model = KNeighborsClassifier()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=42)
```

```python
def test_model(model):
    model.fit(x_train,y_train)
    y_pred = model.predict(x_test)
    print(classification_report(y_test,y_pred))
    print(confusion_matrix(y_test,y_pred))
    
```

```python
test_model(log_model)
```

```python
test_model(knn_model)
```

```python
test_model(svc_model)
```

```python
test_model(random_forest_model)
```

```python
test_model(decision_tree_model)
```

```python
scores = []
for k in range(1,100):
    rfc = RandomForestClassifier(n_estimators=k)
    rfc.fit(x_train,y_train)
    pred_values = rfc.predict(x_test)
    scores.append(accuracy_score(y_test,pred_values))
```

```python
plt.plot(range(1,100),scores)
plt.xlabel('n_estimators')
plt.ylabel('accuracy scores')
plt.xlim(0,100)
# n_estimators = 13
```

```python
random_forest_params = {'criterion':['gini','entropy','log_loss'],
                       'max_depth':list(range(1,10)),
                       }
```

```python
rfc_grid_model = RandomForestClassifier(n_estimators=13)
grid_model = GridSearchCV(rfc_grid_model, param_grid=random_forest_params,cv=10)
grid_model.fit(x_train,y_train)
grid_model.best_params_
```

```python
grid_model.best_score_
```

```python
grid_pred = grid_model.predict(x_test)
print(classification_report(y_test, grid_pred))
confusion_matrix(y_test,grid_pred)
```
